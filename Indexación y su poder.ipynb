## Ahora os vamos a mostrar un problema real que surgió en mongoDB a lo largo de la realización de mi TFG junto con Daniel
### El problema estaba en que teniamos un gran número de datos y a la hora de realizar ciertas operaciones sobre el conjunto sin indexación, las pruebas tardaban horas.

```python
db.Ind.getIndexes()
db.Ind.dropIndexes()
db.Ind.createIndex({player_id:1,year:1})
```

```python
import csv
from pymongo import *
import pymongo
import win_unicode_console
import sys
import  codecs
import time

#CONEXION BASE DE DATOS
mongoClient = MongoClient('localhost',27017)
db = mongoClient.indexacion

#COLECCIONES A TRATAR base de datos
jugadores=db.t_player
norma=db.Ind

#ARCHIVO DE SALIDA
djuga=open("datos_jugadores_cambioTierFila.csv","w",newline='')
salida=csv.writer(djuga, delimiter=',')

juga = jugadores.find({},{'_id':0}).sort("player_id") #de aqui sacamos todos los jugadores

tierAnt='vacio'
datosAnt=[]
datosAct=[]
chivato=0
cont=1
for jugador in juga:
	
	id=jugador['player_id']
	if(cont%100==0):
		print(cont)
	#BUSCAMOS EN LA TABLA NORMALIZADA LAS DIFERENTES TEMPORADAS DE ESE JUGADOR ORDENADAS POR FECHA 
	ju=norma.find({'player_id':id},{'_id':0}).sort("year")
	cabecera=[]
	for a in ju:
		if(a["TIER"]=="NO REGULAR") or a["TIER"]==0 :
			continue
		if tierAnt=='vacio' or tierAnt==a['TIER'] :#caso en el que no ha cambiado de Tier respecto a la temporada anterior
			tierAnt=a['TIER']
			datosAnt=[]# RESETEAMOS LOS DATOS ANTERIORES
			for clave in a :#copiamos todos
				datosAnt.append(a[clave])
			tier_ant=a['TIER']
			score_ant=a['Score']
			if score_ant=="":
				tierAnt='vacio'
				datosAnt=[]
				datosAct=[]
				escribir=[]
				continue
			if(score_ant==0):
				tierAnt='vacio'
				datosAnt=[]
				datosAct=[]
				escribir=[]
				continue	
		else:#CASO EN EL QUE DOS TEMPORADAS SEGUIDAS HA CAMBIADO DE TIER
			
			
			datosAct=[]
			for clave in a :
				if chivato == 0: #Caso añadir cabecera año nuevo
					cabecera.append("bef_"+clave)
				datosAct.append(a[clave])
				tier_act=a['TIER']
				score_act=a['Score']
				if score_act=="":
					score_act=0
			for clave in a :#copiamos todos
				if chivato == 0: #Caso añadir cabecera
					cabecera.append("aft_"+clave)
			if chivato == 0: #Caso añadir cabecera
				cabecera.append("inc_score")
				cabecera.append("inc_tier")
				salida.writerow(cabecera)
				chivato=1
			if score_ant=="":
				tierAnt='vacio'
				datosAnt=[]
				datosAct=[]
				escribir=[]
				continue
			if(score_ant==0):
				tierAnt='vacio'
				datosAnt=[]
				datosAct=[]
				escribir=[]
				continue	
			inc_score=(score_act-score_ant)*100/score_ant
			
			if type(tier_act) is str or type(tier_ant) is str:
				inc_tier=-1
			else:
				inc_tier=tier_act-tier_ant
			escribir=datosAnt+datosAct
			escribir.append(inc_score)
			escribir.append(inc_tier)
			salida.writerow(escribir)
			escribir=[]
			
			#AHORA ACTUALIZAMOS ANTERIOR EL ACTUALIZAMOS
			datosAnt=datosAct
			tierAnt=a['TIER']
	tierAnt='vacio'
	datosAnt=[]
	datosAct=[]
	cont+=1
```
